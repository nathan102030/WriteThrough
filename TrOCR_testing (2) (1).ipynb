# Use TrOCR to turn handwriting to text

```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image

# Load local image
image_path = 'IMG_3948.JPG'  # Replace with your image path
image = Image.open(image_path).convert("RGB")

# Load TrOCR model and processor
processor = TrOCRProcessor.from_pretrained("ndemssie/trocr-finetuned_DeepLRNGCV")
model = VisionEncoderDecoderModel.from_pretrained("ndemssie/trocr-finetuned_DeepLRNGCV")

# Preprocess image
pixel_values = processor(images=image, return_tensors="pt").pixel_values

# Generate text
generated_ids = model.generate(pixel_values)
generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("Generated Text:", generated_text)
```

```python
import nltk
from nltk.util import ngrams
from nltk.corpus import reuters
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter
import random

# Download required NLTK resources
nltk.download('punkt')
nltk.download('reuters')

# Load the Reuters dataset and tokenize into sentences
reuters_text = ' '.join(reuters.words())  # Combine all words into a single text
reuters_sents = [word_tokenize(sent) for sent in sent_tokenize(reuters_text)]  # Tokenize into sentences and words

# Tokenize the entire corpus and generate bigrams and trigrams
tokens = [word.lower() for sentence in reuters_sents for word in sentence]
bigrams = list(ngrams(tokens, 2))  # Bigrams (n=2)
trigrams = list(ngrams(tokens, 3))  # Trigrams (n=3)

# Count the frequency of each bigram and trigram
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

# Function to predict the next word using trigrams (fallback to bigrams if necessary)
def predict_next_word(start_sentence):
    words = word_tokenize(start_sentence)  # Tokenize the input sentence
    if len(words) < 2:
        return predict_using_bigrams(start_sentence)  # Not enough words for trigram prediction

    # Use trigrams if possible
    last_bigram = tuple(words[-2:])  # Last two words as a bigram
    trigram_candidates = {trigram[2]: freq for trigram, freq in trigram_freq.items() if trigram[:2] == last_bigram}

    if trigram_candidates:
        return max(trigram_candidates, key=trigram_candidates.get)  # Most frequent trigram completion

    # Fallback to bigrams
    return predict_using_bigrams(start_sentence)

# Helper function to predict using bigrams
def predict_using_bigrams(start_sentence):
    words = word_tokenize(start_sentence)
    if not words:
        return None  # No input words

    last_word = words[-1].lower()  # Use the last word for bigram prediction
    bigram_candidates = {bigram[1]: freq for bigram, freq in bigram_freq.items() if bigram[0] == last_word}

    if bigram_candidates:
        return max(bigram_candidates, key=bigram_candidates.get)  # Most frequent bigram completion
    return None

# Select 1000 random sentences from Reuters to evaluate the model
random.seed(42)  # For reproducibility
test_data = random.sample(reuters_sents, 1000)

# Evaluate accuracy on the test sentences
def evaluate_accuracy(test_data):
    correct_predictions = 0
    total_predictions = 0

    for sentence in test_data:
        if len(sentence) < 3:
            # Skip sentences too short for trigram prediction
            continue

        # Input string: all but the last word
        input_sentence = ' '.join(sentence[:-1])
        actual_next_word = sentence[-1].lower()  # Convert to lowercase

        predicted_next_word = predict_next_word(input_sentence)

        if predicted_next_word == actual_next_word:
            correct_predictions += 1
        total_predictions += 1

    if total_predictions == 0:
        return 0.0  # Avoid division by zero

    accuracy = correct_predictions / total_predictions
    return accuracy

# Example usage
#start_sentence = "The dictionary is filled with"
generated_text
predicted_word = predict_next_word(generated_text)
print(f"Predicted next word for '{generated_text}': words")

# Evaluate the accuracy
accuracy = evaluate_accuracy(test_data)
print(f"Accuracy on 1000 test sentences from Reuters: {accuracy:.4f}")
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Data extracted from the table
steps = [200, 400, 600, 800, 1000, 1200, 1400]
training_loss = [0.6309, 0.5826, 0.1993, 0.0297, 0.0716, 0.0154, 0.0002]
validation_loss = [0.568761, 0.510521, 0.552562, 0.309079, 0.245185, 0.223296, 0.188567]
cer = [0.213607, 0.159267, 0.223106, 0.092335, 0.072012, 0.061851, 0.051027]
# Plotting data in separate graphs
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Training Loss Plot
axs[0].plot(steps, training_loss, marker='o', color='blue', label='Training Loss')
axs[0].set_title('Training Loss')
axs[0].set_xlabel('Steps')
axs[0].set_ylabel('Loss')
axs[0].grid(True)
axs[0].legend()

# Validation Loss Plot
axs[1].plot(steps, validation_loss, marker='o', color='orange', label='Validation Loss')
axs[1].set_title('Validation Loss')
axs[1].set_xlabel('Steps')
axs[1].set_ylabel('Loss')
axs[1].grid(True)
axs[1].legend()

# CER Plot
axs[2].plot(steps, cer, marker='o', color='green', label='CER (Character Error Rate)')
axs[2].set_title('Character Error Rate (CER)')
axs[2].set_xlabel('Steps')
axs[2].set_ylabel('CER')
axs[2].grid(True)
axs[2].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
